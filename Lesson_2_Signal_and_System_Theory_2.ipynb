### 傅利葉轉換
### 執行以下程式碼並觀察輸出

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
fs = 100
t = np.arange(0, 1, 1/fs)
x = np.cos(2*np.pi*10*t)
plt.figure(figsize=[10,5])
plt.plot(t, x)
plt.title('Signal plot')

xf = np.fft.fft(x)
w_hat = np.arange(0, 2*np.pi, np.pi*2/fs)

plt.figure(figsize=[15,5])

plt.subplot(121)
plt.plot(w_hat, abs(xf))
plt.xlabel(r'$\hat\omega$  (radian)')
plt.title('The magnitude spectrum')

plt.subplot(122)
plt.plot(np.angle(xf))
plt.xlabel(r'$\hat\omega$  (radian)')
plt.title('The phase spectrum')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

fs = 100
t = np.arange(0, 1, 1/fs)
x = np.cos(2*np.pi*10*t)
plt.figure(figsize=[10,5])
plt.plot(t, x)
plt.title('Signal plot')

xf = np.fft.fft(x)
w_hat = np.arange(0, 2*np.pi, np.pi*2/fs)
hat = w_hat * fs /(2 * np.pi)

plt.figure(figsize=[15,5])

plt.subplot(121)
plt.plot(hat, abs(xf))
plt.xlabel(r'f')
plt.title('The magnitude spectrum')

max_magnitude_argmax = np.argmax(abs(xf))
max_magnitude_max = np.max(abs(xf))

print("max:", max_magnitude_max)
print("argmax:", max_magnitude_argmax)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import imageio
from ipywidgets import interact, interactive, fixed
from IPython.display import display
from skimage.transform import rotate

lims = (0, 100, 10)

def edit_image(fs = 100, fn = 10):
    t = np.arange(0, 1, 1/fs)
    x = np.cos(2*np.pi*fn*t)
    plt.figure(figsize=[10,5])
    plt.plot(t, x)
    plt.title('Signal plot')
    xf = np.fft.fft(x)
    w_hat = np.arange(0, 2*np.pi, np.pi*2/fs)
    hat = w_hat * fs / (2 * np.pi)

    plt.figure(figsize=[15,5])

    plt.subplot(121)
    plt.plot(hat, abs(xf))
    plt.xlabel(r'f')
    plt.title('The magnitude spectrum')

    max_magnitude_argmax = np.argmax(abs(xf))
    max_magnitude_max = np.max(abs(xf))

    print("max:", max_magnitude_max)
    print("argmax:", max_magnitude_argmax)

w = interactive(edit_image, fs=lims, fn=lims)
display(w)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

amplitude1 = 1.0  
phase1 = 0.0      
frequency1 = 1.0  

amplitude2 = 0.2 
phase2 = 0.0 
frequency2 = 60

t = np.arange(0, 1, 0.001) 

x1 = amplitude1 * np.cos(2 * np.pi * frequency1 * t + phase1)
x2 = amplitude2 * np.cos(2 * np.pi * frequency2 * t + phase2)


x3 = x1 + x2


plt.figure(figsize=[10, 5], dpi=600)
plt.plot(t, x1, label='X1')
plt.plot(t, x2, label='X2')
plt.plot(t, x3, label='X3 = X1 + X2')
plt.legend()
plt.grid(True)
plt.title('Plot')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```
