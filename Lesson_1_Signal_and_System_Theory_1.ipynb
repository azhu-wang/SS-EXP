問題：上網搜尋 skimage.transform.rotate 的使用方法，將上述介面增加一項新的滑桿。
用這個滑桿可以調整Lena影像的旋轉角度。


```python
!pip install scikit-image
```

```python
import numpy as np
import imageio
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed
from IPython.display import display
from skimage.transform import rotate

lena_img = imageio.imread('lena_std.jpg')

lims = (0.1, 1.0, 0.1)

def edit_image(r=1.0, g=1.0, b=1.0, rotation=0.0):
    lena_rotated = rotate(lena_img, rotation)

    lena_img2 = lena_rotated * 0
    lena_img2[:, :, 0] = r * lena_rotated[:, :, 0]
    lena_img2[:, :, 1] = g * lena_rotated[:, :, 1]
    lena_img2[:, :, 2] = b * lena_rotated[:, :, 2]

    plt.imshow(lena_img2)

w = interactive(edit_image, r=lims, g=lims, b=lims, rotation=(-180.0, 180.0, 1.0))
display(w)
```

## 問題八: 將兩個旋波相加後，並畫在一起。
    
### (1) 產生兩個cos弦波 
<BR>X1: Amplitude: 1, Phase: 0 rad, Frequency: 5 Hz
<BR>X2: Amplitude: 0.7, Phase: pi rad, Frequency: 5 Hz
<BR>X3 = X1 + X2

### (2)就你的觀察X3的頻率為何？

### (3) 使用np.real, np.exp函式，將cos 表示成 complex exponetial 之實部函數，重覆上述動作

### $$\cos(\omega t) = \mathrm{Re}[ e^{j\omega t}] $$



```python
amplitude1 = 1.0  
phase1 = 0.0      
frequency1 = 5.0  

amplitude2 = 0.7 
phase2 = np.pi   
frequency2 = 5.0  

t = np.arange(0, 1, 0.001) 

x1 = amplitude1 * np.cos(2 * np.pi * frequency1 * t + phase1)
x2 = amplitude2 * np.cos(2 * np.pi * frequency2 * t + phase2)


x3 = x1 + x2


plt.figure(figsize=[10, 5], dpi=100)
plt.plot(t, x1, label='X1')
plt.plot(t, x2, label='X2')
plt.plot(t, x3, label='X3 = X1 + X2', linestyle='--')
plt.legend()
plt.grid(True)
plt.title('Two Cosine Waves and Their Sum')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

```python
amplitude1 = 1.0  
phase1 = 0.0      
frequency1 = 5.0  

amplitude2 = 0.7 
phase2 = np.pi   
frequency2 = 5.0  

t = np.arange(0, 1, 0.001)  


x1 = amplitude1 * np.exp(1j * (2 * np.pi  * frequency1 *t +phase1))
x2 = amplitude2 * np.exp(1j * (2 * np.pi  * frequency2 *t +phase2))

x3 = x1 + x2


cosine_exponential_1= np.real(x1)
cosine_exponential_2= np.real(x2)
cosine_exponential_3= np.real(x3)

# 绘制余弦函数的实部
plt.figure(figsize=[10, 5], dpi=100)
plt.plot(t, cosine_exponential_1, label='Real Part1 of e^(j*2π*2t)')
plt.plot(t, cosine_exponential_2, label='Real Part2 of e^(j*2π*2t)')
plt.plot(t, cosine_exponential_3, label='Real Part3 of e^(j*2π*2t)')
plt.grid(True)
plt.legend()
plt.title('Real Part of Cosine Exponential Function')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```
